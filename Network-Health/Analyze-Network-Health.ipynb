```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("log-network-health.log")
df['date'] = pd.to_datetime(df['date'])
```

```python
plt.plot(df.date, df.status)
plt.xlabel('Date')
plt.ylabel('Status')
plt.title('Network monitoring')
plt.xticks(rotation='45')
plt.show()
```
